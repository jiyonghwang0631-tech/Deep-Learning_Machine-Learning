```python
import tensorflow as tf
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.src import datasets, layers, models
fashion_mnist = keras.src.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
test_labels_one_shot = keras.utils.to_categorical(test_labels)
```

```python
model = keras.models.Sequential(name="MNIST_FASHION")
```

```python
model.add(keras.layers.Input((28, 28)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation="relu"))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(10, activation="softmax"))
model.summary()
```

```python
+plt.imshow(train_images[0], cmap='gray')
plt.show()
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(train_images, train_labels_one_shot, epochs=20)
```

```python
(test_loss, test_acc) = model.evaluate(test_images, test_labels_one_shot)
print('정확도:', test_acc)
```
